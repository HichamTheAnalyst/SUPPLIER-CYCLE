```python
Locations_charges_locatives_2023['Compte'] = Locations_charges_locatives_2023['Compte'].astype(str)
Locations_charges_locatives_2024['Compte'] = Locations_charges_locatives_2024['Compte'].astype(str)

Locations_charges_locatives = pd.merge(
    Locations_charges_locatives_2023,
    Locations_charges_locatives_2024,
    on='Compte'
)
```

```python
# Convert both columns to numeric, coercing errors to NaN
Locations_charges_locatives['montans_x'] = pd.to_numeric(Locations_charges_locatives['montans_x'], errors='coerce')
Locations_charges_locatives['montans_y'] = pd.to_numeric(Locations_charges_locatives['montans_y'], errors='coerce')

# Perform the subtraction
Locations_charges_locatives['ecart'] = Locations_charges_locatives['montans_x'] - Locations_charges_locatives['montans_y']
Locations_charges_locatives['percntage'] = (
    (Locations_charges_locatives['montans_x'] - Locations_charges_locatives['montans_y']) /
    Locations_charges_locatives['montans_x'] * 100
).round(2)
Locations_charges_locatives['percntage'] = Locations_charges_locatives['percntage'].apply(lambda x: f"{x:.2f}%")
Locations_charges_locatives
```
